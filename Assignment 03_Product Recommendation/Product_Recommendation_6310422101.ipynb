## **3. Market Basket**

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import re
```

```python
freq_items_set = apriori(Final_df, 
                         min_support=0.08, 
                         use_colnames=True,).sort_values('support', ascending=False).reset_index(drop=True)
#freq_items_set = apriori(Final_df, 
                         #min_support=0.05, 
                         #use_colnames=True,
                         #max_len=2).sort_values('support', ascending=False).reset_index(drop=True)

freq_items_set['total_items'] = freq_items_set['itemsets'].apply(lambda x: len(x))

freq_items_set
```

```python
rules = association_rules(freq_items_set, metric='lift', min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
rules
```

```python
rules['antecedents_length'] = rules['antecedents'].apply(lambda x: len(x))
rules['consequents_length'] = rules['consequents'].apply(lambda x: len(x))
rules
```

```python
#Fitering criteria lift >=2 , support >= 0.1 and confidence >= 0.5 for single item pair
rules_filter = rules[ (rules['antecedents_length'] <= 1) &
                      (rules['consequents_length'] <= 1) &
                      (rules['lift'] >= 1) &
                      (rules['support'] >= 0.1) &
                      (rules['confidence'] >= 0.5) ]
#rules_filter = rules[ (rules['lift'] >= 1) &
                      #(rules['support'] >= 0.2) &
                      #(rules['confidence'] >= 0.5) ]
rules_filter
```

```python
## Check recommendation item
market_basket = rules_filter.sort_values(by=['lift', 'confidence', 'support', 'consequent support', 'antecedent support'], ascending=False)

market_basket.head(20)
```

# **Collaborative Filtering**

```python
#สร้าง Matrix (item x item)
item_item_matrix = pd.DataFrame(index=Final_df.columns,columns=Final_df.columns)
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(Final_df.iloc[:,i],Final_df.iloc[:,j])
```

```python
# unpivot ให้ item กับ item คู่กัน
links = item_item_matrix.rename_axis('related item', 
                                     axis='columns').stack().reset_index()
links.columns = ['item', 'related item','cosine sim']
links.head(20)
```

```python
#filter item ที่ cosine สูงๆ และไม่เอาที่ related กัน
links_filtered=links.loc[ (links['cosine sim'] > 0.1) & 
                         (links['item'] != links['related item']) ]
cf = links_filtered.sort_values(by=['cosine sim'], ascending=False)
```

```python
## Collaborative Filtering (cosine similarity)

cf.head(40)
```

```python
cf.to_csv('Collaborative Filtering.csv')
```
